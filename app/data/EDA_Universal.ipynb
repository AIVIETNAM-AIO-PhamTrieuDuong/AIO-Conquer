# Universal Analyze

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io

def analyze_and_clean_data(data_frame):
    """
    Conducts EDA, cleans data, removes duplicates, and generates statistics.
    Returns: df_clean, num_stats, cat_stats, llm_buffer
    """
    # 1. Duplicate Handling
    initial_count = len(data_frame)
    df_clean = data_frame.drop_duplicates().copy()
    removed_duplicates = initial_count - len(df_clean)
    print(f"### Data Quality Check ###")
    print(f"Total duplicates removed: {removed_duplicates}")

    # 2. Basic Information
    print("\n### Dataset Overview ###")
    print(f"Shape after deduplication: {df_clean.shape}")
    display(df_clean.info())

    # 3. Missing Values
    print("\n### Missing Values ###")
    missing_vals = df_clean.isnull().sum()
    display(missing_vals[missing_vals > 0])

    # 4. Numerical & Categorical Analysis
    num_cols = df_clean.select_dtypes(include=[np.number]).columns
    cat_cols = df_clean.select_dtypes(include=['object', 'bool', 'category']).columns

    print("\n### Numerical Statistics ###")
    display(df_clean.describe())

    if len(num_cols) > 0:
        df_clean[num_cols].hist(figsize=(15, 10), bins=30)
        plt.tight_layout()
        plt.show()

    # 5. Correlation Heatmap
    if len(num_cols) > 1:
        print("\n### Numerical Correlation Heatmap ###")
        plt.figure(figsize=(12, 8))
        sns.heatmap(df_clean[num_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm')
        plt.show()

    # 6. ULTIMATE OFF-DOMAIN ENGINE (LLM Buffer)
    llm_buffer = io.StringIO()
    llm_buffer.write("=== Data Shape ===\n")
    llm_buffer.write(f"Rows: {df_clean.shape[0]}, Columns: {df_clean.shape[1]}\n\n")

    info_summary = pd.DataFrame({
        'DataType': df_clean.dtypes,
        'MissingValues': df_clean.isnull().sum(),
        'UniqueValues': df_clean.nunique()
    })
    llm_buffer.write("=== Missing Values & Data Types ===\n")
    llm_buffer.write(info_summary.to_string() + "\n\n")

    # 7. Cleaning & Stats Extraction
    null_stats = df_clean.isnull().sum().to_dict()
    for col in num_cols:
        df_clean[col] = df_clean[col].fillna(df_clean[col].mean())
    for col in cat_cols:
        df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0] if not df_clean[col].mode().empty else "Unknown")

    num_stats = {}
    # Initialize with empty lists to store column names
    df_clean['outlier_col_name'] = [[] for _ in range(len(df_clean))]

    for col in num_cols:
        Q1, Q3 = df_clean[col].quantile(0.25), df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        is_outlier = (df_clean[col] < (Q1 - 1.5*IQR)) | (df_clean[col] > (Q3 + 1.5*IQR))

        # Append current column name to the list in 'outlier_col_name' for outlier rows
        df_clean.loc[is_outlier, 'outlier_col_name'] = df_clean.loc[is_outlier, 'outlier_col_name'].apply(lambda x: x + [col])

        num_stats[col] = {
            'mean': df_clean[col].mean(),
            'uniques': df_clean[col].nunique(),
            'nulls_fixed': null_stats.get(col, 0),
            'outliers': int(is_outlier.sum()),
            'outlier_perc': (is_outlier.sum()/len(df_clean))*100
        }

    # Post-process: Convert empty lists to "0" for better readability
    df_clean['outlier_col_name'] = df_clean['outlier_col_name'].apply(lambda x: x if len(x) > 0 else "0")

    cat_stats = {}
    for col in cat_cols:
        counts = df_clean[col].value_counts()
        cat_stats[col] = {
            'mode': counts.index[0] if not counts.empty else None,
            'uniques': df_clean[col].nunique(),
            'nulls_fixed': null_stats.get(col, 0),
            'top_5': counts.head(5).to_dict()
        }

    return df_clean, num_stats, cat_stats, llm_buffer

def generate_summary_md(df, num_stats, cat_stats):
    lines = ['# CONQUER QA REPORT: DATA FOOTPRINT\n', '## 1. Numerical Analysis\n']

    for col, s in num_stats.items():
        warning = ""
        if s['outlier_perc'] > 5:
            warning = " ⚠️ **[TRIGGER WARNING: HIGH ANOMALY RATE]**"
        lines.append(f'### {col}\n- **Mean:** {s["mean"]:.2f}\n- **Outliers:** {s["outliers"]} ({s["outlier_perc"]:.2f}%){warning}\n')

    lines.append('\n## 2. Categorical Landscape\n')
    for col, s in cat_stats.items():
        top_str = ", ".join([f"{k} ({v})" for k, v in s['top_5'].items()])
        lines.append(f'### {col}\n- **Mode:** {s["mode"]}\n- **Top 5:** {top_str}\n')

    lines.append(f'\n## 3. System Alerts & Metadata\n')
    total_outlier_rows = len(df[df['outlier_col_name'] != "0"])
    lines.append(f'- **Total Rows with Flagged Anomalies:** {total_outlier_rows}\n')
    lines.append(f'- **Data Integrity:** Specific outlier sources preserved in `outlier_col_name` column.\n')

    text = "".join(lines)
    with open('summary.md', 'w', encoding='utf-8') as f:
        f.write(text)
    return text

# Usage:
# df_clean, n_stats, c_stats, profile = analyze_and_clean_data(df)
```

```python
df_clean, n_stats, c_stats, profile = analyze_and_clean_data(df)

print("\n--- Analysis Complete ---")
print(f"Numerical Stats for {len(n_stats)} columns captured.")
print(f"Categorical Stats for {len(c_stats)} columns captured.")

summary_content = generate_summary_md(df_clean, n_stats, c_stats)
```

# LLM-generated summary markdown file

```python
import google.generativeai as genai


def call_llm_for_insight(summary_text, data_profile, gemini_key):
    """
    Synthesizes semantic insights from statistical summaries using Gemini AI.
    Optimized for Data-Centric QA workflows.
    """
    print("[Conquer] Connecting to AI Synthesis Node...")

    if not gemini_key:
        print("[Error] Gemini API Key is missing. Execution aborted.")
        return

    # 1. AI Synthesis Logic
    try:
        genai.configure(api_key=gemini_key)

        # Priority: 1.5 Flash -> 1.5 Pro -> 2.5 Flash (to manage daily quotas)
        available_models = [m.name for m in genai.list_models()]

        if 'models/gemini-1.5-flash' in available_models:
            target_model = 'gemini-1.5-flash'
        elif 'models/gemini-1.5-pro' in available_models:
            target_model = 'gemini-1.5-pro'
        else:
            target_model = 'gemini-2.5-flash'

        print(f"[Conquer] Optimized Model Selected: {target_model}")
        model = genai.GenerativeModel(target_model)

        prompt = f"""
        Role: Senior Data-Centric AI & QA Specialist.
        Task: Analyze the statistical summary of the Employee Attrition dataset provided below.

        Guidelines:
        1. Analyze Outliers: Explain how these 'system footprints' contribute to QA model robustness.
        2. Bias Detection: Evaluate categorical distributions for potential data imbalance.
        3. Strategic Recommendations: Provide 2 actionable technical steps for the modeling team (Quang and Duong).

        Format: Professional English, concise, and structured in Markdown.

        STATISTICAL REPORT:
        {summary_text}
        The goal is to provide as much detailed information about the dataset features and columns as possible based on these statistics.

        DATASET PROFILE:
        {data_profile.getvalue()}

        Include insights on data distributions, potential outliers, and a description of what each feature likely represents.
        """

        response = model.generate_content(prompt)
        output_md = f"""
---
## 4. LLM Semantic Insights (Engine: {target_model})
> **Status:** Synthesis Successful
> **Node Execution Date:** 2026-04-28

{response.text.strip()}

---
"""
        print(f"[Conquer] AI Insights synthesized successfully via {target_model}.")

    except Exception as e:
        error_msg = str(e)
        print(f"[Warning] AI Node Exception: {error_msg}")
        output_md = f"""
---
## 4. LLM Semantic Insights (Fallback)
**Notice:** AI Insight generation was bypassed due to system constraints.
*Error Log: {error_msg}*
---
"""

    # 2. File Export Logic (Only write once)
    try:
        with open('summary.md', 'a', encoding='utf-8') as f:
            f.write(output_md)
        print("[Conquer] Summary file successfully updated.")
    except IOError as io_err:
        print(f"[Error] Disk I/O Failure: {io_err}")
```

```python
# Step 3–4: Gemini API key — repo .env (GOOGLE_API_KEY or GEMINI_API_KEY), then optional Colab secrets
import os
from pathlib import Path
from dotenv import load_dotenv

def _load_dotenv_from_repo() -> None:
    """Walk up from cwd, load the first .env found. Values from the file replace os.environ entries."""
    for directory in [Path.cwd(), *Path.cwd().resolve().parents]:
        env_path = directory / ".env"
        if not env_path.is_file():
            continue
        for raw in env_path.read_text(encoding="utf-8").splitlines():
            line = raw.strip()
            if not line or line.startswith("#") or "=" not in line:
                continue
            key, _, value = line.partition("=")
            key = key.strip()
            value = value.strip().strip("'\"")
            if key:
                os.environ[key] = value
        break

_load_dotenv_from_repo()
gemini_key = os.environ.get("GOOGLE_API_KEY")

call_llm_for_insight(summary_content, profile, gemini_key=gemini_key)

# Step 4: Export
df_clean.to_csv("conquer_processed_data.csv", index=False)
print("Workflow Success: Enhanced files updated.")
```
